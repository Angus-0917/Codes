```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def critical_max(x,y):
      cp_x,cp_y,cp_i=[],[],[]
      for i in range(len(x)):
        if i+1<len(x) and i-1>=0:
          d1=(y[i+1]-y[i])/(x[i+1]-x[i])
          d2=(y[i]-y[i-1])/(x[i]-x[i-1])
          if d1*d2<0 and y[i]>=y[i+1] and y[i]>=y[i-1]:
            cp_i.append(i)
            cp_x.append(x[i])
            cp_y.append(y[i])
      return cp_i,cp_x,cp_y
```
